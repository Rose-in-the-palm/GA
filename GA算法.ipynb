```python
import numpy as np
import random
import math
import time
```

```python
# 编码函数
def dec2bin(x):
#    判断是否为负数
    flag = 0
    if x<0:
        x=abs(x)
        flag = 1
#     整数部分二进制
    string = bin(int(x))[2:]
#     小数部分二进制
    x -= int(x)
    bins = []
    while x:
        x *= 2
        bins.append('1' if x>=1. else '0')
        x -= int(x)
#   转换为字符串
    a = ''.join(bins)
    return [flag,string+'.'+a]

# 解码函数
def bin2dec(gene):
#    整数部分
    num = int(float(gene[1]))
#   小数部分转换
    decimal = gene[1].split('.')[1]
    for i in range(len(decimal)):
        num += 2**(-i-1)*int(decimal[i])
#   加上正负号
    return -1**gene[0]*num

# 取反字符串指定位置的数
def reverse(string,position):
    string = list(string)
    if string[position]=='.':
        string[position] = "."
    elif string[position] == '0':
        string[position] = "1"
    else:
        string[position] = "0"
    return ''.join(string)
# 判断是否超出范围的函数
def whether_out(gene):
    if -5 <=bin2dec(gene) <= 5:
        return False
    else:
        return True
```

```python
# 目标函数值计算
def f(C):
    return 1/(C[0]**2+C[1]**2+C[2]**2+C[3]**2+1)
# 计算适应值选出best
def Eval(C,scale):
    temp = np.array([])
    for i in range(scale):
        temp = np.append(temp,f(C[i]))
    best = np.argmax(temp)
    return [best,max(temp),temp]
# 选择种群
def select(C,temp,scale):
    selection = np.array([])
    res = temp.sum()
    for i in range(5):
        temp[i] /= res
    for i in range(scale):
        choice = np.random.random()
        if choice<=temp[0]:
            selection = np.append(selection,0)
        elif choice>temp[0] and choice<=(temp[0]+temp[1]):
            selection = np.append(selection,1)
        elif choice>(temp[0]+temp[1]) and choice<=(temp[0]+temp[1]+temp[2]):
            selection = np.append(selection,2)
        elif choice>(temp[0]+temp[1]+temp[2]) and choice<=(temp[0]+temp[1]+temp[2]+temp[3]):
            selection = np.append(selection,3)
        else:
            selection = np.append(selection,4)
#     print(selection)
    for i in range(scale):
        C[i] = C[int(selection[i])].copy()
    
    return selection
# 交配
def copulation(C,selection,scale):
    pc=0.88
    for i in range(scale):
        permit = np.random.random()
        if permit<pc:
            location = np.random.randint(0,3)
            for j in range(location+1,len(C[i])):
                t=C[i][j]
                C[i][j] = C[int(selection[i])][j]
                C[int(selection[i])][j]=t
# 变异
def variation(C,scale):
    pm = 0.05
    for i in range(scale):
        for j in range(4):
            gene = dec2bin(C[i][j])
            for k in range(len(gene[1])):
                p = np.random.random()
                if p<0.05 and gene[1][k]!='.':
                    gene[1] = reverse(gene[1],k)
                    ran=k
#                     print('gene:',gene)
                    while whether_out(gene):
                        gene[1] = reverse(gene[1],ran)                          
                        ran = np.random.randint(0,len(gene[1])-1)
                        gene[1] = reverse(gene[1],ran)
            C[i][j] = bin2dec(gene)
```

```python
def GA(length,scale):
    for i in range(scale):
        for j in range(4):
            C[i][j]=np.random.uniform(-5,5)
    for j in range(length):
        result = Eval(C,scale)
        selection = select(C,result[2],scale)
        copulation(C,selection,scale)
        variation(C,scale)
        result = Eval(C,scale)
#         print(result[1],C[result[0]],j)
        if result[1]==1.:
            break
    return [result[1],C[result[0]],i]
```

```python
start=time.time()
# length:迭代次数，N:种群规模
length,N=1000,30
C = np.zeros((N,4))
# GA(length,scale)
print(GA(length,N))
end = time.time()
print("GA算法运行时间:%.2f s"%(end-start))
```
